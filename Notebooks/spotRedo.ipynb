```python
#Created this for getting the year from the spotify api
import pandas as pd
import numpy as np
import config
from sqlalchemy import create_engine
import tekore as tk
```

```python
client_id = config.spot_id
client_secret = config.spot_secret
#tekore creds
app_token = tk.request_client_token(client_id, client_secret)
```

```python
spotify = tk.Spotify(app_token,chunked_on=True)

```

```python
spotify_data = "../Resources/cleaned_spotify.csv"
spotify_df = pd.read_csv(spotify_data)
spotify_df.head()
```

```python
songyear = []

sp9 = list(spotify_df['track_id'].iloc[0:9000])
sp18 = list(spotify_df['track_id'].iloc[9001:18000])
sp27 = list(spotify_df['track_id'].iloc[18001:27000])
sp36 = list(spotify_df['track_id'].iloc[27001:36000])
sp45 = list(spotify_df['track_id'].iloc[36001:45000])
sp54 = list(spotify_df['track_id'].iloc[45001:54000])
sp63 = list(spotify_df['track_id'].iloc[54000:63000])
sp72 = list(spotify_df['track_id'].iloc[63001:72000])
sp81 = list(spotify_df['track_id'].iloc[72001:81000])
sp90 = list(spotify_df['track_id'].iloc[81001:90000])
sp99 = list(spotify_df['track_id'].iloc[90001:99000])
sp108 = list(spotify_df['track_id'].iloc[99001:108000])
sp117 = list(spotify_df['track_id'].iloc[108001:11700])
sp126 = list(spotify_df['track_id'].iloc[117001:126000])
sp135 = list(spotify_df['track_id'].iloc[126001:135000])
sp144 = list(spotify_df['track_id'].iloc[135001:144000])
sp153 = list(spotify_df['track_id'].iloc[144001:153000])
sp162 = list(spotify_df['track_id'].iloc[153001:162000])
sp171 = list(spotify_df['track_id'].iloc[162001:171000])
sp176 = list(spotify_df['track_id'].iloc[171001:176774])
```

```python
#function to go through sets
def datehunt(lset):
        dates = spotify.tracks(lset,market='US')
        return dates
```
