```python
#Created this for cleaning the the data set of any nulls, as well as duplicate IDs
import pandas as pd
import numpy as np
import config
from sqlalchemy import create_engine
```

```python
#Load Spotify Data
spotify_data = "../Resources/SpotifyFeatures.csv"
spotify_df = pd.read_csv(spotify_data)

spotify_df
```

```python
#Removed Duplicate ids and rows with nulls
spotifyDup_df = spotify_df.drop_duplicates(subset=['track_id'])
spotifyCleaned_df = spotifyDup_df.dropna()
spotifyCleaned_df = spotifyCleaned_df.rename(columns={'artist_name':'artist','track_name':'track'})
spotifyCleaned_df
```

```python
# Correcting values with difference in quotes
spotifyCleaned_df.loc[(spotifyCleaned_df.genre == 'Children’s Music'), 'genre'] = "Children's Music"
```

```python
spotifyCleaned_df['genre'].unique()
```

```python
#Saves to new csv
spotifyCleaned_df.to_csv('../Resources/cleaned_spotify.csv')
```
